##### identifying the data that is in the first dataframe but not in the second, based on categories of a column

```python
#finding inconsistent categories

#establishing a list of items that is df_1 but not in df_2, ie. the difference
inconsistent_categories = set(df_1['common_column']).difference(df_2['common_column'])

#rows for this difference in categories (ie. the rows pertaining to the inconsistency)
inconsistent_rows = df_1['common_column'].isin(inconsistent_categories)

#subset the data based on these rows
inconssisten_data = df_1[inconsistent_rows]
```

##### dropping these differences in data

```python
consistent_data = df_1[~inconsistent_rows]
#using the delta symbol while subsetting which returns everything except the inconsistent data
```
